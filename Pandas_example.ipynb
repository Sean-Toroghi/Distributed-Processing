__Filter rows and columns of a DataFrame1 to just show the txId and price__

[[ ]] creates a copy of the DataFrame with only the specified columns

```python
dataframe1[['txId','price']].head(5)
```

__Filter rows based on a criteria__

```python
dataframe1[dataframe1['price']< 3].head(10)
```

__Calculate sum and average of a column__

```python
sum_price = dataframe1['price'].sum()
```

```python
sum_price
```

```python
mean_price = dataframe1['price'].mean()
```
